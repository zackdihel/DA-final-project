## Part 2: Data + Identification Strategy

### Research Design

- **Research question:** Does higher education attainment cause a lower probability of unemployment?
- **Identification strategy:** IV
- **Key assumption:** Exclusion Restriction: instrumental variable affects unemployment *only* through education. Instrumental variable will also be uncorrelated with unobserved predictors.
- **Treatment variable:** Education (college degree)
- **Outcome variable:** Employment status (unemployment)
- **Controls:** Age, race, sex, income, year, industry, ethnicity
- **Why prediction alone is insufficient:** Does not identify biases and causalities that can lead to the incorrect conclusions being drawn. IV will analyze what happens when education changes, and how that impacts unemployment.

## Part 3: Analysis

### 3a. Naive Estimate (Biased Benchmark)

This simple comparison is expected to be biased. Document *why*.

```python
# --- Naive OLS ---
X_naive = df[['college_degree']]
X_naive = sm.add_constant(X_naive)
y = df['unemployed']
naive_model = sm.OLS(y, X_naive).fit()
print(naive_model.summary())
naive_estimate = naive_model.params['college_degree']
naive_ci = naive_model.conf_int().loc['college_degree'].values
print(f'\nNaive estimate: {naive_estimate:.4f} (95% CI: [{naive_ci[0]:.4f}, {naive_ci[1]:.4f}])')
```

**Why the naive estimate is biased:** Naive estimate is biased because it is an example of the omitted variable bias. There are variables that coexist with whether someone has a college degree that also help to explain their employment status; this effect is not being captured by the naive model. There are observed effects that a college-degree has on unemployment rate, but the unobserved effects are being included as well which is likely causing downward bias overstating negative effects on the estimate shown above.

### 3b. Causal Estimate

```python
# --- Causal Method: Instrumental Variables ---
df = pd.read_csv('../data/educ_cleaned_instr.csv')

df = df[[c for c in df.columns if not c.startswith('yr_')]]
year_dummies = pd.get_dummies(df['YEAR'], prefix='yr', drop_first=True).astype(int)
df = pd.concat([df, year_dummies], axis=1)
year_cols = year_dummies.columns.tolist()

# Define variables
treatment  = 'college_degree'
instrument = 'tuition_per_fte'
outcome    = 'unemployed'
controls   = ['AGE', 'SEX', 'RACE', 'HISPAN', 'is_MA'] + year_cols
model_df = df[[outcome, treatment, instrument] + controls].dropna()

# First stage
X_fs    = sm.add_constant(df[[instrument] + controls])
fs      = sm.OLS(df[treatment], X_fs).fit(cov_type='HC1')
college_degree_hat = fs.fittedvalues

fs_coef  = fs.params[instrument]
fs_se    = fs.bse[instrument]
fs_tstat = fs.tvalues[instrument]
fs_pval  = fs.pvalues[instrument]
fs_fstat = fs.fvalue

print("=" * 55)
print("FIRST STAGE: tuition_per_fte → college_degree")
print("=" * 55)
print(f"Coefficient : {fs_coef:.6f}")
print(f"Std Error   : {fs_se:.6f}")
print(f"T-Statistic : {fs_tstat:.3f}")
print(f"P-Value     : {fs_pval:.4f}")
print(f"F-Statistic : {fs_fstat:.2f}\n")

if fs_fstat > 104:
    print(f"  F = {fs_fstat:.2f} > 104 — strong instrument (Lee et al. 2022) ✓")
elif fs_fstat > 10:
    print(f"  F = {fs_fstat:.2f} > 10  — relevant instrument (Stock-Yogo) ✓")
else:
    print(f"  F = {fs_fstat:.2f} < 10  — WEAK INSTRUMENT WARNING ✗")
print()
print("Interpretation:")
print(f"  A $1 increase in tuition per FTE changes college degree")
print(f"  attainment probability by {fs_coef:.6f} percentage points.")
```

A more intuitive interpretation of above results: a $1000 increase in tuition would lead to a -0.5% change in degree attainment.

```python
#Second stage
X_ss = sm.add_constant(
    pd.concat([college_degree_hat.rename('college_degree_hat'),
               model_df[controls]], axis=1)
)
ss = sm.OLS(model_df[outcome], X_ss).fit(cov_type='HC1')
ss_coef = ss.params['college_degree_hat']
ss_ci   = ss.conf_int().loc['college_degree_hat'].values
print(f"\n--- Second Stage (manual 2SLS) ---")
print(f"Estimate : {ss_coef:.4f}")
print(f"95% CI   : [{ss_ci[0]:.4f}, {ss_ci[1]:.4f}]")

Y = model_df[outcome]
X = sm.add_constant(model_df[controls])
D = model_df[[treatment]]
Z = model_df[[instrument]]

iv = IV2SLS(dependent=Y, exog=X, endog=D, instruments=Z).fit(cov_type='robust')

iv_coef = iv.params[treatment]
iv_ci   = iv.conf_int().loc[treatment].values
iv_se   = iv.std_errors[treatment]
iv_pval = iv.pvalues[treatment]

X_ols   = sm.add_constant(model_df[[treatment] + controls])
ols     = sm.OLS(model_df[outcome], X_ols).fit(cov_type='HC1')
ols_coef = ols.params[treatment]
ols_ci   = ols.conf_int().loc[treatment].values

print(f"\n--- IV 2SLS via linearmodels (correct SEs) ---")
print(f"Estimate : {iv_coef:.4f}")
print(f"Std Error: {iv_se:.4f}")
print(f"95% CI   : [{iv_ci[0]:.4f}, {iv_ci[1]:.4f}]")
print(f"P-Value  : {iv_pval:.4f}")
```

### 3c. Prediction Model (for comparison)

```python
# --- Predictive Model (NOT causal — for comparison only) ---
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import roc_auc_score, log_loss

RANDOM_STATE = 42
all_features = ['college_degree', 'SEX', 'AGE', 'INCTOT']

rf = RandomForestClassifier(n_estimators=100, random_state=RANDOM_STATE, n_jobs=-1)

# predict_proba gives probability of unemployment (class 1)
# using a subsample, because our N would take a very long time to execute
df_sample = df.sample(n=100000, random_state=42)
y_pred_proba = cross_val_predict(rf, df_sample[all_features],
                                  df_sample['unemployed'],
                                  cv=5, method='predict_proba')[:, 1]

print(f'Prediction AUC-ROC : {roc_auc_score(df_sample["unemployed"], y_pred_proba):.3f}')
print(f'Prediction Log-Loss: {log_loss(df_sample["unemployed"], y_pred_proba):.3f}')
print('\nNote: This tells us how well we can PREDICT the outcome,')
print('but NOT how the treatment CAUSES changes in the outcome.')
```

### 3e. Robustness Check

```python
# --- Robustness: IV model ---
# Setup cell — run this before the robustness script
import pandas as pd
import numpy as np
import statsmodels.api as sm
from linearmodels.iv import IV2SLS

df = pd.read_csv('../data/educ_cleaned_instr.csv')
df = df[[c for c in df.columns if not c.startswith('yr_')]]
year_dummies = pd.get_dummies(df['YEAR'], prefix='yr', drop_first=True).astype(int)
df = pd.concat([df, year_dummies], axis=1)
year_cols = year_dummies.columns.tolist()

outcome    = 'unemployed'
treatment  = 'college_degree'
instrument = 'tuition_per_fte'
###

def run_iv(data, endog_var, instrument_var, control_vars, label):
    d = data[[outcome, endog_var, instrument_var] + control_vars].dropna()
    iv = IV2SLS(
        dependent   = d[outcome],
        exog        = sm.add_constant(d[control_vars]),
        endog       = d[[endog_var]],
        instruments = d[[instrument_var]]
    ).fit(cov_type='robust')
    X_fs = sm.add_constant(d[[instrument_var] + control_vars])
    fs = sm.OLS(d[endog_var], X_fs).fit(cov_type='HC1')
    fs_fstat = fs.fvalue

    coef = iv.params[endog_var]
    ci   = iv.conf_int().loc[endog_var].values
    pval = iv.pvalues[endog_var]
    return {'label': label, 'coef': coef, 'ci': ci,
            'pval': pval, 'fstat': fs_fstat, 'n': len(d)}

results = {}

controls_base = ['AGE', 'SEX', 'RACE', 'HISPAN', 'is_MA'] + year_cols
results['Baseline IV'] = run_iv(df, treatment, instrument, controls_base, 'Baseline IV')

df_norecession = df[~df['YEAR'].isin([2008,2009,2020])].copy()
yr_cols_norecession = [c for c in year_cols if c not in ['yr_2008', 'yr_2009', 'yr_2020']]
controls_norecession = ['AGE', 'SEX', 'RACE', 'HISPAN', 'is_MA'] + yr_cols_norecession
results['Drop Recessions'] = run_iv(df_norecession, treatment, instrument, controls_norecession, 'Drop Recessions')

results['Alt. Instrument (is_MA)'] = run_iv(df, treatment, 'is_MA', ['AGE', 'SEX', 'RACE', 'HISPAN']+year_cols,'Alt. Instrument (is_MA)')

controls_min = ['AGE', 'SEX', 'is_MA'] + year_cols
results['Minimal Controls'] = run_iv(df,treatment,instrument,controls_min,'Minimal Controls')

df_ind = df[df['IND'] !=0].copy()
df_ind = df_ind[[c for c in df_ind.columns if not c.startswith('ind_')]]
ind_dummies = pd.get_dummies(df_ind['IND'],prefix='ind',drop_first=True).astype(int)

ind_dummies = ind_dummies.loc[:, ind_dummies.sum() > 100]

df_ind = pd.concat([df_ind, ind_dummies],axis=1)
ind_cols = ind_dummies.columns.tolist()
controls_exp = ['AGE', 'SEX', 'RACE', 'HISPAN', 'is_MA'] + year_cols + ind_cols
results['Expanded Controls'] = run_iv(df_ind,treatment,instrument,controls_exp,'Expanded Controls')

print("=" * 110)
print("IV ROBUSTNESS CHECKS")
print("=" * 110)
print(f"{'Specification':<30} {'Estimate':>10} {'95% CI':>26} {'P-Val':>8} {'F-Stat':>8} {'N':>10}")
print("-" * 110)
for key, r in results.items():
    sig = '*' if r['pval'] < 0.05 else ''
    print(f"{r['label']:<30} {r['coef']:>10.4f}   "
          f"[{r['ci'][0]:>7.4f}, {r['ci'][1]:>7.4f}]   "
          f"{r['pval']:>6.4f} {r['fstat']:>8.2f} {r['n']:>10,}")
print("=" * 110)
print("* p < 0.05")
print("\nInterpretation guide:")
print("  - Estimates stable across specs → result is robust")
print("  - F-stat > 10 in all specs      → instrument remains relevant")
print("  - CI crossing zero in all specs → consistent insignificance")
```

## Part 4: Threats to Identification

**Minimum 500 words. Be honest — this is where you demonstrate critical thinking.**

### 1. Most Serious Threat

- **Threat:** Violation of Exclusion Restriction
- **Direction of bias:** Biased towards zero or even positive; understating true negative causal effect
- **What would address it:** Using an instrument that is purely supply-side, rather than one determined by fiscal conditions. Alternatively, making use of any potential natural experiments that are based around sharp changes in tuition policies.

### 2. Second Threat

- **Threat:** Limitations resulting form a Two-State sample (MA and NH)
- **Why it matters:** Conditions surrounding both college decisions and labor markets vary from state to state; different populations' variations may lead to different results.
- **Partial mitigation:** Partially mitigated this by using a 24 year sample saize and using fixed effects to help absorb macroeconomic trends. Additional mitigation could be achieved by expanding the sample size beyond just MA and NH to include other New England states, which should improve the external validity without compromising the general setup of the study.

### 3. What I Cannot Rule Out

Cannot rule out that the true causal effect of a college degree on unemployment is zero, or negligibly small. The IV estimate of -0.0944 is statistically insignificant. Inherent to using an instrumental variable study, precision was sacrificed in favor of causal credibility. Also, it cannot be ruled out that the value of a college degree has shifted in such a way as to be less meaningful when it comes to employability. If degrees have become less proof of acquired skills/specializations and become more of a signaling mechanism, our study will not be able to capture that shift.
